```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
from google.colab import files
import cv2
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
cols = ['AGE',	'GENDER',	'SMOKING',	'FINGER_DISCOLORATION',	'MENTAL_STRESS',	'EXPOSURE_TO_POLLUTION'	,'LONG_TERM_ILLNESS',	'ENERGY_LEVEL',	'IMMUNE_WEAKNESS',	'BREATHING_ISSUE',	'ALCOHOL_CONSUMPTION',	'THROAT_DISCOMFORT',	'OXYGEN_SATURATION',	'CHEST_TIGHTNESS'	,'FAMILY_HISTORY',	'SMOKING_FAMILY_HISTORY',	'STRESS_IMMUNE','PULMONARY_DISEASE']
med = pd.read_csv("/content/Lung Cancer Dataset.csv", header=None, names=cols)
med['PULMONARY_DISEASE'] = med['PULMONARY_DISEASE'].map({ 'YES': 1, 'NO': 0})
med.drop(0, inplace=True)
numeric_cols = cols
med[numeric_cols] = med[numeric_cols].apply(pd.to_numeric, errors='coerce')
med
```

```python

x = med.drop('PULMONARY_DISEASE', axis=1)
y = med.PULMONARY_DISEASE

```

```python
test_part = 0.15
train_part = 1 - test_part

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=train_part, random_state=42)
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
X_test_cat = to_categorical(X_test)

```

```python
model = Sequential()
model.add(Dense(600, activation = 'relu', input_shape = (X_train.shape[1],)))
model.add(Dense(400, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(2,activation = 'softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train_cat, epochs = 40, batch_size = 1, verbose = 1, validation_split = 0.2)
```

```python

loss, acc = model.evaluate(X_test, y_test_cat)
print(f"Точность: {acc}")
```

```python

predictions = model.predict(X_test)
y_pred = predictions.argmax(axis=1)
y_ans = y_test_cat.argmax(axis=1)
for id_prediction in range(10):
  print(f'Ответ модели(вероятность 0 и вероятность 1): {np.asarray(predictions)[id_prediction]}')
  print(f'Ответ модели: {np.asarray(y_pred)[id_prediction]}')
  print(f'Правильный ответ: {np.asarray(y_ans)[id_prediction]}')
  print('')
model.save('model_LC.h5')
```
